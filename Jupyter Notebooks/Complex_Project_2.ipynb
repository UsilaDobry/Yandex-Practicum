### Описание данных
- `Rougher feed` — исходное сырье
- `Rougher additions` (или `reagent additions`) — флотационные реагенты: Xanthate, Sulphate, Depressant
- `Xanthate` — ксантогенат (промотер, или активатор флотации);
- `Sulphate` — сульфат (на данном производстве сульфид натрия);
- `Depressant` — депрессант (силикат натрия).
- `Rougher process` (англ. «грубый процесс») — флотация
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота

### Параметры этапов
- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size` — размер гранул сырья
- `feed rate` — скорость подачи

### Наименование признаков
*Наименование признаков должно быть такое*: `[этап].[тип_параметра].[название_параметра]` - Пример: `rougher.input.feed_ag`
*Возможные значения для блока [этап]*:
- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

*Возможные значения для блока [тип_параметра]*:
- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

### Целевые признаки
- эффективность обогащения чернового концентрата `rougher.output.recovery`;
- эффективность обогащения финального концентрата `final.output.recovery`.

### Формула расчёта эффективности обогощения

$$Recovery = \frac{C * (F - T)}{F * (C - T)} * 100$$

где:
C — доля золота в концентрате после флотации/очистки;
F — доля золота в сырье/концентрате до флотации/очистки;
T — доля золота в отвальных хвостах после флотации/очистки.

### Формула метрики качества

$$sMAPE = \frac{1}{N} * \sum_{i=1}^{N}\frac{|y_i-\hat{y_i}|}{(|y_i|+|\hat{y_i}|)/2} * 100$$

### Формула итоговой метрики

$$Final-sMAPE = 0.25 * sMAPE(rougher) + 0.75 * sMAPE(final)$$

### Проверка правильности вычисления эффективности обогащения

```python
def recovery_efficiency(row):
    '''
    Функция принмает на вход объект (строка)
    рассчитывает и возвращает эффективность восстановления
    :param row:
    :return:
    '''
    coef_recovery = (((row['rougher.output.concentrate_au']) * ((row['rougher.input.feed_au']) - (row['rougher.output.tail_au'])))/
                       (((row['rougher.input.feed_au']) * ((row['rougher.output.concentrate_au']) - (row['rougher.output.tail_au']))))) * 100
    return coef_recovery
# вычислим эффективность восстановления на обучающей выборке
recovery_train = df_train.dropna(subset = ['rougher.input.feed_au',
                                           'rougher.output.concentrate_au',
                                           'rougher.output.tail_au',
                                           'rougher.output.recovery'],axis = 0).apply(recovery_efficiency, axis = 1)
# вычислим среднюю абсолютную ошибку между нашими расчетами и данными
print('Средняя абсолютная ошибка =', mae(df_train.dropna(subset = ['rougher.input.feed_au',
                                    'rougher.output.concentrate_au',
                                    'rougher.output.tail_au',
                                    'rougher.output.recovery'])['rougher.output.recovery'], recovery_train))
```

### Вывод
Ошибка очень мала. Эффективность обогощения в данных рассчитана корректно.

### Расчет суммарной концентрации на разных этапах

```python
def func_total_concentarution(row):
    '''
    Функция рассчитывает суммарную концентрацию по всем компонентам
    на вход принмает объект (строку) датафрейма
    :param row:
    Функция возвращает серию значений сумм концентрации компонентов на каждом этапе
    :return:
    '''
    # сформируем список этапов обработки золотой руды
    list_steps = ["rougher","primary_cleaner","final"]
    # выборка значений сырья
    input_feed_au = row["rougher.input.feed_au"]
    input_feed_ag = row["rougher.input.feed_ag"]
    input_feed_pb = row["rougher.input.feed_pb"]
    input_feed_sol = row["rougher.input.feed_sol"]
    # списки концентрации продукта на этапах: флотации, первичной обработки, вторичной обработки
    out_rougher = []
    out_primary_cleaner = []
    out_final = []
    list_arrays =[out_rougher,out_primary_cleaner,out_final]
    for step,array in zip(list_steps,list_arrays):
        array.append(row[step+".output.concentrate_au"])
        array.append(row[step+".output.concentrate_ag"])
        array.append(row[step+".output.concentrate_pb"])
        array.append(row[step+".output.concentrate_sol"])
    # суммирование концентрации компонентов на каждом этапе
    sum_before_steps = input_feed_ag+input_feed_au+input_feed_pb+input_feed_sol
    sum_rougher = sum(out_rougher)
    sum_primary = sum(out_primary_cleaner)
    sum_final = sum(out_final)
    return pd.Series([sum_before_steps,sum_rougher,sum_primary,sum_final])

total_concetrate = df_train.apply(func_total_concentarution,axis = 1)
total_concetrate.columns = ["before_rougher","rougher","primary_cleaner","final"]

display(total_concetrate.head())
```

## Модель
Для поиска оптимальной модели для прогнозирования используем один из инструментов автоматического машинного обучения **TROT**, который рассматривает несколько алгоритмов машинного обучения (случайные леса, линейные модели, SVM и т. д.) В конвейере с несколькими шагами предварительной обработки (импутация данных, масштабирование, PCA, выбор функций и т. д.) он рассчитывает гиперпараметры для всех моделей и этапов предварительной обработки, а также предлагает несколько способов ансамбля алгоритмов в конвейере.
В задаче используем значения параметров **TROT**, значительно меньше значений по умолчанию, в целях ускорения расчетов. Параметр кросс-валидации используем по умолчанию `cv=5`.
### Функция расчета метрики sMAPE

```python
def func_smape(y, y_pred):
    '''
    Функция принимает на вход реальные метки и предсказаниями
    :param df_rez:
    Функция возвращает оценочную метрику sMAPE
    :return: score
    '''
    score = np.mean(abs(y - y_pred) / ((abs(y) + abs(y_pred)) / 2)) * 100
    return score

def func_final_smape(score_1, score_2):
    '''
    Функция рассчитывает итоговую оценочную метрику
    на вход принимает оценки по двум этапам (rougher, final)
    :param score_1:
    :param score_2:
    Функция возвращает итоговую оценку
    :return: score
    '''
    score = 0.25 * score_1 + 0.75 * score_2
    return score
```

### Моделирование для этапа флотации

```python
# введем оценочную метрику sMAPE в алгоритм оптимизации TROT
scoring = make_scorer(func_smape, greater_is_better=False)
# выполиним моделирование для этапа флотации
tpot_rougher = TPOTRegressor(generations=3, population_size=5, verbosity=2, n_jobs=-1, random_state=12345, scoring=scoring)
# обучение модели с учителем
tpot_rougher.fit(X_train_rougher_std, target_rougher_train)
```
