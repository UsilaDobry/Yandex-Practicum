## Предобработка данных
### Работа с пропусками

```python
# переименуем поле cityCenters_nearest
df = df.rename(columns={'cityCenters_nearest':'city_centers_nearest'})
```

```python
# подсчёт количества пропущенных значений по каждому полю (столбцу)
print('Количество пропусков:', df.isna().sum(), sep='\n')
```

##### Поле `ceiling_height` (высота потолка)
имеет 9195 пропусков, что составляет 38.8%

```python
# посмотрим уникальные значения
df['ceiling_height'].value_counts()
```

Поле имеет аномально низкие (до 1 м) и аномально высокие (до 100 м) значения.
Значения можно округлить до одно знака после запятой.
Принять пограничные значения: минимально 2,5м, максимально 6м. остальные значения заменить на 0.
Пропущенные значения также заменить на 0.

```python
# меняем на ноль значения за границами допустимого диапозона
df.loc[df['ceiling_height'] > 6, 'ceiling_height'] = 0
df.loc[df['ceiling_height'] < 2.5, 'ceiling_height'] = 0
# округляем значения до одной цифры после запятой
df['ceiling_height'] = round(df['ceiling_height'], 1)
# меняем пропущенные значения на ноль
df['ceiling_height'] = df['ceiling_height'].fillna(0)
# проверяем результат
df['ceiling_height'].value_counts().sort_values(ascending=False)
```

##### Поле `floors_total` (всего этажей)
имеет 86 пропусков, что составляет 0,4%, возможно информация отсутствует, меняем их на ноль

```python
#df['floors_total'] = df['floors_total'].fillna(0)
```

##### Поле `locality_name` (название населенного пункта)
имеет 49 пропусков, что составляет 0,2%, возможно информация отсутствует, меняем их на пустое значение

```python
df['locality_name'] = df['locality_name'].fillna('')
```

```python
# проверим уникальные значения на неявные дубликаты
df['locality_name'].value_counts()
```

```python
# напишем функцию для группировки неявных дубликатов
def locality_name_group(str):
    '''
    функция принимает строку, сплитует её на слова в список
    в цикле перебираем слова в списке, если слово начинается с заглавной буквы
    то меняем его на пустое значение
    функция возвращает модифицированную строку
    '''
    str_list = str.split()
    for i in range(len(str_list)):
        if str_list[i].istitle() == False:
            str_list[i] = ''
    return ' '.join(str_list).lstrip()

# в исходном датафрейме изменяем значение поля `locality_name`
df['locality_name'] = df['locality_name'].apply(locality_name_group)
# проверим внесенные изменения
print(df['locality_name'].value_counts())
```

##### Поле `living_area` (жилая площадь)
имеет 1903 пропуска, что составляет 8%, возможно информация отсутствует, меняем на ноль

```python
df['living_area'] = df['living_area'].fillna(0)
```

##### Поле `is_apartment` (это апартаменты)
имеет 20924 пропусков, что составляет 88.3%, в описании это значение типа `bool`.
строгих технических требований для классификации апартаментов нет.

#### Зависимость стоимости объекта от категории этажности

```python
# создадим сводную таблицу, группируя по трем категориям этажности (первый\последний\другой) с агрегацией стоимости объектов по среднему значению.
df2 = df.pivot_table(index='floor_category', values='last_price', aggfunc='mean')
print('Сводная таблица по среднему значению стоимости:', df2, sep='\n')
# создадим сводную таблицу, группируя по трем категориям этажности (первый\последний\другой) с агрегацией стоимости объектов по медианному значению.
df2 = df.pivot_table(index='floor_category', values='last_price', aggfunc='median')
print('Сводная таблица по медианному значению стоимости:', df2, sep='\n')
```


Зависимость по среднему значению:
- Объекты, расположенные в категории "другой", имеют самую высокую среднюю стоимость.
- Объекты, расположенные на первом этаже, имеют самую низкую среднюю стоимость.
Зависимость по медианному значению:
- Объекты, расположенные в категории "другой" (между первым и последним этажами), имеют самое высокое медианное значение стоимости.
- Объекты, расположенные на первом этаже имеют самое низкое медианное значение стоимости.

#### Зависимость cтоимости объекта от даты размещения: дня недели, месяца и года.

```python
# создадим новыe датафреймы из исходного, группировкой по соотвествующим признакам и агрегацией по средней цене объекта
df_price_weekday_mean = df.groupby('weekday_exposition', as_index=False)['last_price'].mean()
df_price_weekday_count = df.groupby('weekday_exposition', as_index=False)['last_price'].count()
df_price_month = df.groupby('month_exposition', as_index=False)['last_price'].mean()
df_price_year = df.groupby('year_exposition', as_index=False)['last_price'].mean()

plt.rcParams['figure.figsize'] = 30, 15
plt.rcParams['axes.titlesize'] = 20
# Две строки, два столбца. Текущая ячейка - 1
pylab.subplot (2, 2, 1)
# строим столбчатую диаграмму
pylab.bar(x=df_price_weekday_mean['weekday_exposition'], height=df_price_weekday_mean['last_price'])
pylab.title('Зависимость цены от дня недели')
# Две строки, два столбца. Текущая ячейка - 2
pylab.subplot (2, 2, 2)
# строим  столбчатую диаграмму
pylab.bar(x=df_price_weekday_count['weekday_exposition'], height=df_price_weekday_count['last_price'])
pylab.title('Зависимость количества объявлений от дня недели')
# Две строки, два столбца. Текущая ячейка - 3
pylab.subplot(2, 2, 3)
# строим столбчатую диаграмму
pylab.bar(x=df_price_month['month_exposition'], height=df_price_month['last_price'])
pylab.title('Зависимость цены от месяца')
# Две строки, два столбца. Текущая ячейка - 4
pylab.subplot(2, 2, 4)
# строим столбчатую диаграмму
pylab.bar(x=df_price_year['year_exposition'], height=df_price_year['last_price'])
pylab.title('Зависимость цены от года')
```

Зависимость стоимости объекта от даты размещения объявления:
- Зависимость средней цены объекта от дня недели размещения объявления не ярко выражена. Объявления, размещенные в субботу имеют самую низкую цену, объявления, размещенные во вторник имеют самую высокую среднюю цену объекта. Эта зависимость коррелирует с активностью пользователей в социальных сетях. Наибольшая активность в соцсетях со вторника по четверг, наименьшая на выходных днях. Это видно на графике зависимости количества объявлений от дня недели.
- Зависимость средней цены объекта от месяца размещения объявления более интересное. Мы видим сезонное движение цены, повышение  цены приходится на апрель и сентябрь, понижение цены приходится на июнь.
- Зависимость средней цены объекта от года размещения объявления в целом говорит нам о падении цены на недвижимость с 2014 года. В 2017, 2018 годах рынок недвижимости стагнировал, в 2019 году отмечается рост цен.
