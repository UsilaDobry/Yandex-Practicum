# Отток клиентов
Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком.

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

```python
# data preprocessing
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
from sklearn.model_selection import RandomizedSearchCV
from sklearn.utils import shuffle
from sklearn.decomposition import PCA

# data visualization
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt

# Algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

## Исследование задачи
### Разделение данных на выборки
Произведем разбивку данных в пропорциях:
- обучающие - 3 части
- валидационные - 1 часть
- тестовые - 1 часть

```python
target = df['Exited']
features = df.drop(['Exited'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=12345, stratify=target)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.25, random_state=12345, stratify=y_train)

print(X_train.shape)
print(y_train.shape)
print(X_valid.shape)
print(y_valid.shape)
print(X_test.shape)
print(y_test.shape)
```

### Стандартизация несбалансированных данных

```python
sc = StandardScaler()
# вычислим параметры μ (эмпирическое среднее) и σ (стандартное отклонение) для каждой размерности признаков из тренировочных данных
sc.fit(X_train)
# стандартизируем данные, используя для этого расчетные параметры μ и σ.
X_train_std = sc.transform(X_train)
X_valid_std = sc.transform(X_valid)
X_test_std = sc.transform(X_test)
```

### Предварительное обучение на несбалансированных данных
#### Логистическая регрессия

```python
%%time
model_logr = LogisticRegression(C=1.0, random_state=12345)
model_logr.fit(X_train_std, y_train)
# выполним прогноз на валидационных данных
y_pred_logr = model_logr.predict(X_valid_std)
# оценим модель
f1_logr = f1_score(y_valid, y_pred_logr)
acc_logr = accuracy_score(y_valid, y_pred_logr)
roc_auc_logr = roc_auc_score(y_valid, y_pred_logr)
```

#### Логистическая регрессия в версии со стахостическим градиентным спуском

```python
%%time
model_logr_sgd = SGDClassifier(loss='log', random_state=12345)
model_logr_sgd.fit(X_train_std, y_train)
# выполним прогноз на валидационных данных
y_pred_logr_sgd = model_logr_sgd.predict(X_valid_std)
# оценим модель
f1_logr_sgd = f1_score(y_valid, y_pred_logr_sgd)
acc_logr_sgd = accuracy_score(y_valid, y_pred_logr_sgd)
roc_auc_logr_sgd = roc_auc_score(y_valid, y_pred_logr_sgd)
```

#### Перцептрон

```python
%%time
model_ppn = Perceptron(max_iter=40, eta0=0.1, random_state=12345)
model_ppn.fit(X_train_std, y_train)
# выполним прогноз на валидационных данных
y_pred_ppn = model_ppn.predict(X_valid_std)
# оценим модель
f1_ppn = f1_score(y_valid, y_pred_ppn)
acc_ppn = accuracy_score(y_valid, y_pred_ppn)
roc_auc_ppn = roc_auc_score(y_valid, y_pred_ppn)
```

#### Перцептрон в версии со стахостическим градиентным спуском

```python
%%time
model_ppn_sgd = SGDClassifier(loss='perceptron', random_state=12345)
model_ppn_sgd.fit(X_train_std, y_train)
# выполним прогноз на валидационных данных
y_pred_ppn_sgd = model_ppn_sgd.predict(X_valid_std)
# оценим модель
f1_ppn_sgd = f1_score(y_valid, y_pred_ppn_sgd)
acc_ppn_sgd = accuracy_score(y_valid, y_pred_ppn_sgd)
roc_auc_ppn_sgd = roc_auc_score(y_valid, y_pred_ppn_sgd)
```

#### Машины опорных векторов в версии со стахостическим градиентным спуском

```python
%%time
model_svm_sgd = SGDClassifier(loss='hinge', random_state=12345)
model_svm_sgd.fit(X_train_std, y_train)
# выполним прогноз на валидационных данных
y_pred_svm_sgd = model_svm_sgd.predict(X_valid_std)
# оценим модель
f1_svm_sgd = f1_score(y_valid, y_pred_svm_sgd)
acc_svm_sgd = accuracy_score(y_valid, y_pred_svm_sgd)
roc_auc_svm_sgd = roc_auc_score(y_valid, y_pred_svm_sgd)
```

#### Дерево решений

```python
%%time
model_dtc = DecisionTreeClassifier(random_state=12345)
model_dtc.fit(X_train_std, y_train)
# выполним прогноз на валидационных данных
y_pred_dtc = model_dtc.predict(X_valid_std)
# оценим модель
f1_dtc = f1_score(y_valid, y_pred_dtc)
acc_dtc = accuracy_score(y_valid, y_pred_dtc)
roc_auc_dtc = roc_auc_score(y_valid, y_pred_dtc)
```

#### Случайный лес

```python
%%time
model_rfc = RandomForestClassifier(n_estimators=100, random_state=12345)
model_rfc.fit(X_train_std, y_train)
# выполним прогноз на валидационных данных
y_pred_rfc = model_rfc.predict(X_valid_std)
# оценим модель
f1_rfc = f1_score(y_valid, y_pred_rfc)
acc_rfc = accuracy_score(y_valid, y_pred_rfc)
roc_auc_rfc = roc_auc_score(y_valid, y_pred_rfc)
```

#### Метод к-ближайших соседей

#### Увеличение выборки положительного класса 1

```python
def upsample(features, target, repeat):
    '''
    Функция принимает на вход
    :param features: - признаки
    :param target: - метки
    :param repeat: - кратность увеличения положительного класса
    Функция возращает
    :return: - преобразованные признаки и метки
    '''
    # разделим признаки и метки на отрицательные и положительные значения
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    # увеличим положительные значения, затем объединим с отрицательными значениями
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    # перемешаем преобразованные данные
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)

    return features_upsampled, target_upsampled
# преобразуем тренировочные данные с помощью функции upsample()
features_upsampled, target_upsampled = upsample(X_train, y_train, 4)
# выведим размер преобразованных данных для обучения
print(features_upsampled.shape)
print(target_upsampled.shape)

class_frequency = target_upsampled.value_counts(normalize=True)
print(class_frequency)
# построим график баланса классов
plt.rcParams['figure.figsize'] = 10, 5
plt.title('Баланс классов')
class_frequency.plot(kind='bar')
plt.show()
```

### Настройка гиперпараметров Случайного леса
Сначала для черновой настройки используем алгоритм RandomizedSearchCV, который позволяет довольно грубо исследовать широкие диапазоны значений.

Задаем значения гиперпараметров через генератор списков, затем создаем словарь `param_dist`, содержащий, для каждого гиперпараметра, диапазон значений, которые нужно испытать. Далее, инициализируем объект rs с помощью функции `RandomizedSearchCV()`, передавая ей нашу модель Случайный лес `model_rfc`, `param_dist`, число итераций и число кросс-валидаций, которые нужно выполнить.

Гиперпараметр `verbose` позволяет управлять объёмом информации, который выводится моделью в ходе её работы (наподобие вывода сведений в процессе обучения модели). Гиперпараметр `n_jobs` позволяет указывать то, сколько процессорных ядер нужно использовать для обеспечения работы модели. Установка `n_jobs` в значение -1 приведёт к более быстрой работе модели, так как при этом будут использоваться все ядра процессора.

Мы будем заниматься подбором следующих гиперпараметров:

- `n_estimators` — число «деревьев» в «случайном лесу».
- `max_features` — число признаков для выбора расщепления.
- `max_depth` — максимальная глубина деревьев.
- `min_samples_split` — минимальное число объектов, необходимое для того, чтобы узел дерева мог бы расщепиться.
- `min_samples_leaf` — минимальное число объектов в листьях.
- `bootstrap` — использование для построения деревьев подвыборки с возвращением.
- `criterion` - вид неоднородности.

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num = 10)]
criterion = ['gini', 'entropy']
max_features = ['log2', 'sqrt']
max_depth = [int(x) for x in np.linspace(start = 1, stop = 15, num = 15)]
min_samples_split = [int(x) for x in np.linspace(start = 2, stop = 50, num = 10)]
min_samples_leaf = [int(x) for x in np.linspace(start = 2, stop = 50, num = 10)]
bootstrap = [True, False]
param_dist = {'n_estimators': n_estimators,
              'criterion': criterion,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf,
              'bootstrap': bootstrap}
rs = RandomizedSearchCV(model_rfc,
                        param_dist,
                        n_iter = 100,
                        cv = 3,
                        verbose = 1,
                        n_jobs=-1,
                        random_state=0)
rs.fit(features_upsampled, target_upsampled)
rs.best_params_
```
